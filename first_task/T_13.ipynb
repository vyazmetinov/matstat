```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import scipy.optimize as opt
import seaborn as sns
import math as mt
from skimage.io import imread, imsave

np.random.seed(42)
plt.style.use('ggplot')
```

```python
def get_s_from_sigma(sigma, n):
    return n/(n-1)*sigma**2

alpha = 0.05

sgm_egypt_length = 5.722
sgm_egypt_width = 4.612
sgm_europe_length = 6.161
sgm_europe_width = 5.055
n = 139
m = 1000
s_egypt_length = get_s_from_sigma(sgm_egypt_length, n)
s_egypt_width = get_s_from_sigma(sgm_egypt_width, n)
s_europe_length = get_s_from_sigma(sgm_europe_length, m)
s_europe_width = get_s_from_sigma(sgm_europe_width, m)

# статистика критерия
delta_wave_length = s_egypt_length/s_europe_length
delta_wave_width = s_egypt_width/s_europe_width

left_q = st.f.ppf(alpha/2, n-1, m-1) # левая квантиль
right_q = st.f.ppf(1-alpha/2, n-1, m-1) # правая квантиль

_, ax = plt.subplots(1, 1, figsize=(12, 6))
x = np.linspace(0.001, 5, 1000)
y = 1 - st.f.cdf(right_q, n-1, m-1, scale=x) + st.f.cdf(left_q, n-1, m-1, scale=x)
ax.plot(x, y, color='red')
ax.set_title('График Мощности')
ax.set_xlabel("theta")
ax.set_ylabel("W(theta)")
plt.show()
```
